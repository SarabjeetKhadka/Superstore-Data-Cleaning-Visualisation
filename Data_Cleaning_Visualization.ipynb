```python
# 6. Import Libraries

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

# Display Settings
pd.set_option('display.max_columns', None)

# Visualization Settings
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

print("Libraries Imported Successfully")
```

```python
# 7. Load Dataset

# df = pd.read_csv('../data/Superstore.csv')
df = pd.read_csv('/content/Superstore.csv')

print("Dataset Loaded Successfully")
```

```python
# 8. Initial Data Exploration
#   - Shape
#   - Columns
#   - Data Types
#   - Summary Statistics

# Dataset Overview

print("Dataset Shape:", df.shape)

print("\nRows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nMemory Usage:")
print(df.memory_usage(deep=True).sum()/1024**2, "MB")

# Number of Unique Values per Column

df.nunique().sort_values(ascending=False)


# Numerical and Categorical Features

numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = df.select_dtypes(include='object').columns.tolist()

print("Numerical Columns:")
print(numerical_cols)

print("\nCategorical Columns:")
print(categorical_cols)

# Statistical Summary

df.describe().T

# Categorical Summary

df.describe(include='object').T
```

```python
# 9. Data Quality Assessment

# - Missing Values
df.isnull().sum()

missing_percentage = (df.isnull().sum()/len(df))*100
plt.show()
print(missing_percentage)

sns.heatmap(df.isnull())
plt.show()

# - Duplicate Records
  #* Check Duplicate Records
duplicate_count = df.duplicated().sum()
print(f"Total Duplicate Records: {duplicate_count}")
# View Duplicate Records
duplicates = df[df.duplicated()]
duplicates.head()

# - Data Type Issues
df.dtypes
df.info()
df[['Order Date','Ship Date']].head()



# - General Data Consistency

# *Check Empty Strings
(df == '').sum()

# *Check Leading and Trailing Spaces
text_columns = df.select_dtypes(include='object').columns

for col in text_columns:
    inconsistent = (df[col].astype(str) != df[col].astype(str).str.strip()).sum()

    print(f"{col}: {inconsistent}")

# *Negative Sales Validation
negative_sales = df[df['Sales'] < 0]
print("Negative Sales Records:", len(negative_sales))

# *Negative Profit Validation
negative_profit = df[df['Profit'] < 0]
print("Negative Profit Records:", len(negative_profit))

# *Discount Range Check
df['Discount'].describe()

invalid_discount = df[
    (df['Discount'] < 0) |
    (df['Discount'] > 1)
]

print("Invalid Discount Records:", len(invalid_discount))
```

```python
# 10. Data Cleaning
#    - Missing Values
for col in numerical_cols:
    df[col].fillna(df[col].median(), inplace=True)
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)


#    - Duplicate Records
df.drop_duplicates(inplace=True)


#    - Data Type Corrections
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

#    - General Data Consistency
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

#    - Standardizing Formatting
df.columns = df.columns.str.strip()

text_columns = [
    'City',
    'State',
    'Country',
    'Region',
    'Market',
    'Category',
    'Sub-Category'
]

for col in text_columns:
    df[col] = df[col].str.strip()
    df[col] = df[col].str.title()

#    - Data Validation
df = df[df['Sales'] >= 0]
df = df[df['Profit'] >= 0]
df = df[df['Discount'] >= 0]
df = df[df['Discount'] <= 1]

#    -Feature Engineering
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

```

```python
# 11. Outlier Detection
""" Outliers are observations that differ significantly from the majority of the data.

In retail datasets, outliers may occur due to:

- High-value customer purchases
- Bulk corporate orders
- Seasonal demand spikes
- Data entry errors

Therefore, outliers must be carefully analyzed before deciding whether they should be treated or retained.

This section uses:

1. Boxplots for visualization
2. Interquartile Range (IQR) Method for detection
3. Business interpretation before treatment decisions """

# Numerical Columns for Outlier Analysis

numerical_columns = [
    'Sales',
    'Profit',
    'Quantity',
    'Discount',
    'Shipping Cost'
]

df[numerical_columns].describe().T

#    - Boxplots for Numerical Features

plt.figure(figsize=(15,8))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2,3,i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()


#    - IQR Method

outlier_summary = []

for col in numerical_columns:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    outliers = df[
        (df[col] < lower_bound) |
        (df[col] > upper_bound)
    ]

    outlier_summary.append([
        col,
        round(Q1,2),
        round(Q3,2),
        round(IQR,2),
        len(outliers)
    ])

outlier_df = pd.DataFrame(
    outlier_summary,
    columns=[
        'Feature',
        'Q1',
        'Q3',
        'IQR',
        'Number of Outliers'
    ]
)

outlier_df.sort_values(by='Number of Outliers', ascending=False)


# Percentage of Outliers

for col in numerical_columns:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    outliers = df[
        (df[col] < lower_bound) |
        (df[col] > upper_bound)
    ]

    percentage = (len(outliers) / len(df)) * 100

    print(
        f"{col}: {len(outliers)} outliers "
        f"({percentage:.2f}%)"
    )

    # Visualizing Outliers Individually

for col in numerical_columns:

    plt.figure(figsize=(8,4))

    sns.boxplot(x=df[col])

    plt.title(f'Outlier Detection - {col}')

    plt.show()


"""
## Outlier Analysis Summary

### Observation

The IQR method identified outliers in several numerical variables such as Sales, Profit, Quantity, Discount, and Shipping Cost.

### Interpretation

These outliers represent observations that lie significantly outside the normal range of values.

In a retail business environment, such observations may correspond to:

- Large corporate purchases
- High-value customers
- Seasonal sales spikes
- Premium products

### Business Impact

Removing all outliers may eliminate valuable business information and distort the true performance of the business.

For example:

- High sales transactions contribute significantly to revenue.
- High profits may identify top-performing products.
- High shipping costs may indicate important logistics patterns.

### Treatment Decision

Outliers will be retained in the dataset because there is no evidence that they are data entry errors or invalid transactions.

Retaining these observations preserves important business information and provides a more realistic view of retail operations."""
```

```python
# 14 EXECUTIVE DASHBOARD CODE - C (visualization)


# EXECUTIVE DASHBOARD


fig, axes = plt.subplots(
    2,
    2,
    figsize=(18,12)
)


# Regional Performance


region_sales = (
    df.groupby('Region')['Sales']
    .sum()
    .sort_values(ascending=False)
)

axes[0,0].bar(
    region_sales.index,
    region_sales.values
)

axes[0,0].set_title('Regional Sales Performance')

axes[0,0].set_xlabel('Region')

axes[0,0].set_ylabel('Sales')

axes[0,0].tick_params(axis='x', rotation=20)

# Monthly Sales Trend

monthly_sales = (
    df.groupby('Order Month Name')['Sales']
    .sum()
    .reindex([
        'January',
        'February',
        'March',
        'April',
        'May',
        'June',
        'July',
        'August',
        'September',
        'October',
        'November',
        'December'
    ])
)

axes[0,1].plot(
    monthly_sales.index,
    monthly_sales.values,
    marker='o'
)

axes[0,1].set_title('Monthly Sales Trend')

axes[0,1].tick_params(axis='x', rotation=45)

# Profit by Category

category_profit = (
    df.groupby('Category')['Profit']
    .sum()
)

axes[1,0].bar(
    category_profit.index,
    category_profit.values
)

axes[1,0].set_title('Profit by Category')

axes[1,0].set_xlabel('Category')

axes[1,0].set_ylabel('Profit')

# Top Products

top_products = (
    df.groupby('Product Name')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

axes[1,1].barh(
    top_products.index,
    top_products.values
)

axes[1,1].set_title('Top 10 Products by Sales')

plt.tight_layout()

plt.show()
```
